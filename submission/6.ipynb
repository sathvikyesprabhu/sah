```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop, sgd , Adam
import matplotlib.pyplot as plt 
from sklearn.cross_validation import train_test_split
```

```python
data_tr=pd.read_csv("manipal_sah_tr.csv")
data_test=pd.read_csv("manipal_sah_test.csv")
data_tr
```

```python


data_tr["Outcome at 3 months"] = np.where(data_tr["Outcome at 3 months"]>1, 0, data_tr["Outcome at 3 months"])
#data_test["Outcome at Discharge"] = np.where(data_test["Outcome at Discharge"]>1, 0, data_test["Outcome at Discharge"])

#data_tr["Outcome at 3 months"] = np.where(data_tr["Outcome at 3 months"]<=3, 0, data_tr["Outcome at 3 months"])
#data_tr["Outcome at 3 months"] = np.where(data_tr["Outcome at 3 months"]>3, 1, data_tr["Outcome at 3 months"])
#data_test["Outcome at Discharge"] = np.where(data_test["Outcome at Discharge"]<=3, 0, data_test["Outcome at Discharge"])
#data_test["Outcome at Discharge"] = np.where(data_test["Outcome at Discharge"]>3, 1, data_test["Outcome at Discharge"])

x_train=data_tr[data_tr.columns[0:5]]
y_train=data_tr[data_tr.columns[8]]
x_test=data_test[data_test.columns[0:5]]
y_test=data_test[data_test.columns[7]]


batch_size = 32
num_classes = 2
epochs = 25

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)

#y_train=y_train[:,1:]
#y_test=y_test[:,1:]
#x_train = x_train.astype('float32')
#x_test = x_test.astype('float32')

#np.std(x_train[x_train.columns[0]])
y_train
```

```python
df=x_train
x_train=(df-df.mean())/(df.std())
x_test=(x_test-df.mean())/(df.std())
x_test
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.33, random_state=7)
Y_train.shape
```

```python
model = Sequential()
#model.add(Dense(15, activation='relu', input_shape=(5,)))
#model.add(Dense(10,activation='relu'))
model.add(Dense(num_classes, activation='softmax' ,input_shape=(5,)))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])

history = model.fit(X_train, Y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=2,
                    validation_data=(X_test, Y_test))
y_test = model.predict(x_test, verbose=0)
```

```python
result=np.argmax(y_test,axis=1)
result
```
